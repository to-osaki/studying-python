```python
import matplotlib.pyplot as plt
import matplotlib.style
matplotlib.style.use('ggplot')
matplotlib.style.available
```

```python
fig, axis = plt.subplots(2) # １つのfigureに対して複数のsubplotを指定できる
axis[0].set_title("1")
axis[1].set_title("2")
axis[0].plot([1, 2, 3], [4, 5, 6], linestyle = "--", color = "aqua")
axis[1].plot([1, 2, 3], [2, 2, 5], linestyle = "-.", linewidth = 3)
fig.suptitle("title")
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(123)
mu = 100
sigma = 15
x = np.random.normal(mu, sigma, 1000)
fig, ax = plt.subplots()

font_dict = {
    "family": 'monospace',
    "size": 10,
    "weight": "light"
}

ax.set_xlabel('xlabel', family = 'fantasy', size = 15, weight = 'bold')
ax.set_ylabel('ylabel', fontdict = font_dict, size = 20) # 辞書のパラメータは上書き可
ax.text(80, 100, "hoge", fontdict = font_dict) # 任意のテキスト

n, bins, patches = ax.hist(x, bins=5, orientation='horizontal') # 標準ではbins=10
for i, num in enumerate(n):
    print('{:.2f} - {:.2f} {}'.format(bins[i], bins[i + 1], num))
```

```python
import matplotlib.pyplot as plt 
labels = ['spam', 'ham', 'egg'] 
x = [10, 3, 1] 
fig, ax = plt.subplots() 
ax.pie(x, labels=labels, startangle=90, counterclock=False, shadow=True, autopct='%1.2f%%', explode = [0, 0.5, 0.1]) 
ax.axis("equal") # アスペクト比を固定
```
